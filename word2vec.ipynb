```python
import gensim.downloader as api
```

```python
word2vec = api.load("word2vec-google-news-300")      #a big 1.5GB model, takes time in downloading
```

```python
word2vec["scooter"]
```

```python
word2vec.similarity("scooter", "bicycle")
```

```python
word2vec.most_similar("scooter", topn = 15) #topn = 10 by default
```
