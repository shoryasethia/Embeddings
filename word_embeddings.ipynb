```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
import tensorflow as tf
import numpy as np
```

```python
senetnces = [ "I ate my apple yeasterday.",
             "That day what were you doing under shades.",
             "I am a good developer.",
             "The cup of tea.",
             "Prime minister is appointed for third time in a row.",
             "If you like this, give it a star."
]
```

```python
senetnces
```

```python
vocab_size = 10000
```

```python
one_hot?
```

```python
one_hot_vector = [one_hot(word, vocab_size) for word in senetnces]
one_hot_vector
```

```python
max_sentence_length = 10

padded_one_hot_rep = pad_sequences(one_hot_vector, padding = 'pre', maxlen = max_sentence_length)
```

```python
padded_one_hot_rep
```

```python
from keras import layers, models

input_layer = layers.Input(shape = (max_sentence_length,), name = 'input_layer')
output_layer = layers.Embedding(vocab_size, output_dim = 10, name = 'output_layer')(input_layer)

model = models.Model(input_layer, output_layer, name = 'word_embedding_model')

model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.predict(padded_one_hot_rep)
```
